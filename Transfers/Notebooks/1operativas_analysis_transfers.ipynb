```python
#correlation matrix - pearson coefficient
transfersd.corr()                          #IBS, ITI, MBW - -3.4e-5, -1.54e-4, -8.3e-4
```

```python
#correlation matrix - kendall coefficient
transfersd.corr(method = 'kendall')               #IBS, ITI, MBW - -3.4e-5, -1.54e-4, -8.3e-4
```

```python
#correlation matrix - spearman coefficient
transfersd.corr(method = 'spearman')         #IBS, ITI, MBW - -3.4e-5, -1.54e-4, -8.3e-4
```

```python
#covariance matrix - positive/negative - variables positively/negatively related // high/low - strong/weak relationship
#IBS, ITI, MBW - -1.187800e-9, -2.357123e-8, -6.685491e-7
transfersd.cov()
```

```python
#chi, mi and f-value conditions

#chi
#if chi_p > 0.05 - null hypothesis - variables independent
#if chi_p <= 0.05 - alternate hypothesis - variables dependent

#mi
#if mi = 0 - variables independent
#if mi > 0 - variables dependent

#f-value
#if f_p <= 0.05 - alternate hypothesis - variables dependent
#if f_p > 0.05 - null hypothesis variables - independent
```

```python
#TRFIBS - all conditions verified - eliminate
from sklearn.feature_selection import f_classif, chi2, mutual_info_classif

X, y = transfersd[['is_fraud']], transfersd[['operativa_TRFIBS']]

chi2_score, chi_2_p_value = chi2(X,y)
f_score, f_p_value = f_classif(X,y)
mut_info_score = mutual_info_classif(X,y)

df = pd.DataFrame({"chi2": chi2_score, "chi2_p": chi_2_p_value, "F": f_score, "F_p": f_p_value, "MI": mut_info_score})
df 
```

```python
#TRFITI - all conditions verified - eliminate
from sklearn.feature_selection import f_classif, chi2, mutual_info_classif

X, y = transfersd[['is_fraud']], transfersd[['operativa_TRFITI']]

chi2_score, chi_2_p_value = chi2(X,y)
f_score, f_p_value = f_classif(X,y)
mut_info_score = mutual_info_classif(X,y)

df = pd.DataFrame({"chi2": chi2_score, "chi2_p": chi_2_p_value, "F": f_score, "F_p": f_p_value, "MI": mut_info_score})
df 
```

```python
#TRFMBW - all conditions are not verified - do not eliminate
from sklearn.feature_selection import f_classif, chi2, mutual_info_classif

X, y = transfersd[['is_fraud']], transfersd[['operativa_TRFMBW']]

chi2_score, chi_2_p_value = chi2(X,y)
f_score, f_p_value = f_classif(X,y)
mut_info_score = mutual_info_classif(X,y)

df = pd.DataFrame({"chi2": chi2_score, "chi2_p": chi_2_p_value, "F": f_score, "F_p": f_p_value, "MI": mut_info_score})
df  
```

```python
#after operativas analysis there was the conclusion to eliminate operativas TRFITI e TRFIBS
target_col = 'operativa'

transfers_opr = transfers[(transfers[target_col] == 'TRFINT') | (transfers[target_col] == 'TRFIPS') | 
                          (transfers[target_col] == 'TRFITC') | (transfers[target_col] == 'TRFMBW') |
                          (transfers[target_col] == 'TRFSEP')].copy()
```
