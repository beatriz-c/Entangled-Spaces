```python
#original frauds before synthetic generation
import pandas as pd

fraudsb = pd.read_csv(r'C:/Users/BeatrizCarvalho/OneDrive - Closer Consultoria Lda/Documents/Entangled-Spaces/Payments/Datasets/0frauds_payments_nd_corr.csv')
```

```python
#how many frauds existed originally
len(fraudsb)
```

```python
#reduced amount of original nonfrauds 
nfraudsb = pd.read_csv(r'C:/Users/BeatrizCarvalho/OneDrive - Closer Consultoria Lda/Documents/Entangled-Spaces/Payments/Datasets/0nonfrauds_payments_nd_corr.csv', 
                       nrows = 10000)
```

```python
#concat dataframe
ub_sample = pd.concat([fraudsb, nfraudsb], ignore_index = True)
```

```python
#class ratios
ub_sample['is_fraud'].value_counts()/ub_sample.shape[0]
```

```python
#plot
import seaborn as sns
import matplotlib.pylab as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

#variables
X = ub_sample[['amount', 'accountbalance', 'hour', 'week']]
y = ub_sample['is_fraud']

#scale data before PCA
pca = PCA()
pipe = Pipeline([('scaler', StandardScaler()), ('pca', pca)])
Xt = pipe.fit_transform(X)

#plot
sns.scatterplot(x = Xt[:, 0], y = Xt[:, 1], c = y, style = y, linewidth = 0.25, size = y, hue = y, sizes = {0: 50, 1:150},
                palette = {0: 'Blue', 1: 'Red'})

#legend
plt.xlabel('Component 1', fontsize = 14)
plt.ylabel('Component 2', fontsize = 14)
plt.yticks(fontsize = 13)
plt.xticks(fontsize = 13)
plt.legend(title = '')

#zoom
plt.xlim([-1, 5])
plt.ylim([-2, 3])

#save
plt.savefig(r'C:/Users/BeatrizCarvalho/OneDrive - Closer Consultoria Lda/Documents/Entangled-Spaces/Payments/Profiles/imbalanced_distribution.png')
```
