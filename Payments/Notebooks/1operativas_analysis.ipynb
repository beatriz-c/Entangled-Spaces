```python
#chi, mi and f-value conditions

#chi
#if chi_p > 0.05 - null hypothesis - variables independent
#if chi_p <= 0.05 - alternate hypothesis - variables dependent

#mi
#if mi = 0 - variables independent
#if mi > 0 - variables dependent

#f-value
#if f_p <= 0.05 - alternate hypothesis - variables dependent
#if f_p > 0.05 - null hypothesis variables - independent
```

```python
#RTFITC - all conditions verified - eliminate
from sklearn.feature_selection import f_classif, chi2, mutual_info_classif

X, y = paymentsd[['is_fraud']], paymentsd[['operativa_RTFITC']]

chi2_score, chi_2_p_value = chi2(X,y)
f_score, f_p_value = f_classif(X,y)
mut_info_score = mutual_info_classif(X,y)

df = pd.DataFrame({"chi2": chi2_score, "chi2_p": chi_2_p_value, "F": f_score, "F_p": f_p_value, "MI": mut_info_score})
df 
```
