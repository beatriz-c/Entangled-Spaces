```python
#data loading
import dask.dataframe as dd

dtype = {'trusted_indicator': 'object'}
transfers = dd.read_csv('0transfers_corr.csv', dtype = dtype)
```

```python
#visualize the whole dataframe
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
#correlation matrix for all columns except for the entity and reference columns
transfers.corr().compute()
```

```python
#keep only the fraud and operativas columns
del transfers[transfers.columns[0]] 

transfers = transfers.drop(['canal', 'entity', 'reference', 'trusted_indicator', 'iban_orig', 'iban_dest', 'amount', 
                            'accountbalance', 'ipaddress', 'is_tablet', 'is_pc', 'is_mobile', 'is_touch', 'clientid', 
                            'browser_family', 'os_family', 'hour', 'week', 'timestamp'], axis = 1) 
```

```python
#covariance matrix 
#positive number if the variables are positively related and negative number if they are negatively related
#high covariance indicates strong relationship and low value means a weak relationship
transfers.cov().compute()
```

```python
#make sure that there's only 2 columns left
transfers.shape
```

```python
#name of the columns left
transfers.shape
```

```python
#create dummies for operativa to get correlations
transfersd = dd.get_dummies(transfers.categorize(), columns = ['operativa']).compute()
```

```python
#correlation matrix - pearson coefficient
transfersd.corr()                          #IBS, ITI, MBW - -3.4e-5, -1.52e-4, -8.4e-4
```

```python
#correlation matrix - kendall coefficient
transfersd.corr(method = 'kendall')               #IBS, ITI, MBW - -3.4e-5, -1.52e-4, -8.4e-4
```

```python
#correlation matrix - spearman coefficient
transfersd.corr(method = 'spearman')         #IBS, ITI, MBW - -3.4e-5, -1.52e-4, -8.4e-4
```

```python
#chi, mi and f-value for operativas with no frauds

#chi
#if chi_p > 0.05 null hypothesis variables independent
#if chi_p <= 0.05 alternate hypothesis variables dependent

#mi
#if mi = 0 independent
#if mi > 0 higher dependency

#f-value
#if f_p <= 0.05 alternate hypothesis variables dependent
#if f_p > 0.05 null hypothesis variables independent

from sklearn.feature_selection import f_classif, chi2, mutual_info_classif

X, y = transfersd[['is_fraud']], transfersd[['operativa_TRFIBS']]

chi2_score, chi_2_p_value = chi2(X,y)
f_score, f_p_value = f_classif(X,y)
mut_info_score = mutual_info_classif(X,y)

df = pd.DataFrame({"chi2": chi2_score, "chi2_p": chi_2_p_value, "F": f_score, "F_p": f_p_value, "MI": mut_info_score})
df 
#verifica tudo posso eliminar TRFIBS  #david - the warning
```
